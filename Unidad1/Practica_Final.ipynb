```python
from sklearn.linear_model import LinearRegression
from scipy.stats import pearsonr
import numpy as np

X1 = df1[["HorasEstudio"]]
y1 = df1[["Calificacion"]]

X2 = df2[["HorasEstudio"]]
y2 = df2[["Calificacion"]]

X3 = df3[["HorasEstudio"]]
y3 = df3[["Calificacion"]]

modelo1 = LinearRegression()
modelo2 = LinearRegression()
modelo3 = LinearRegression()

modelo1.fit(X1, y1)
modelo2.fit(X2, y2)
modelo3.fit(X3, y3)

pendiente = modelo1.coef_[0][0]
intercepto = modelo1.intercept_[0]

print(f"Ecuación Grupo: y = {pendiente:.2f}x + {intercepto:.2f}")


corr1, _ = pearsonr(df1["HorasEstudio"], df1["Calificacion"])
corr2, _ = pearsonr(df2["HorasEstudio"], df2["Calificacion"])
corr3, _ = pearsonr(df3["HorasEstudio"], df3["Calificacion"])

print(f"Coeficiente de correlación de Pearson Grupo 1: {corr1:.3f}")
print(f"Coeficiente de correlación de Pearson Grupo 2: {corr2:.3f}")
print(f"Coeficiente de correlación de Pearson Grupo 3: {corr3:.3f}")
```
