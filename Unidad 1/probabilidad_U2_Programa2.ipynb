```python
nombres = ["a", "b", "c", "d", "e", "f", "g"]
n = len(nombres)
r = 3

Pr = mt.factorial(n) / (mt.factorial(n - r))

print("Existen ", Pr, " posibles permutaciones sin repeticin del total de ",
      n, " nombres ", " en grupos de ", r, " en ", r)

print("Aqu las permutaciones")
permutaciones_nombres = list(permutations(nombres, r))

for permutation in permutaciones_nombres:
    print(permutation)
```
